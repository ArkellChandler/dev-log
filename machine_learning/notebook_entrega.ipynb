```python
# @title 🚀 Pipeline ETL - Santander Dev Week (TOTVS Edition)
# @markdown Execute esta célula para instalar as dependências e rodar o motor de dados.

import pandas as pd
import json

# 1. Simulação de Extração (Mock de Dados para o Colab)
data = {
    'ID': [1, 2, 3],
    'Cliente': ['Arkell', 'TOTVS Player', 'Dev User'],
    'Score': [95, 88, 72]
}
df = pd.DataFrame(data)

# 2. Transformação (Lógica de Negócio)
df['Status'] = df['Score'].apply(lambda x: 'Premium' if x > 90 else 'Standard')

# 3. Carga (Gerando o JSON que o Dashboard PHP leria)
result_json = df.to_json(orient='records', indent=4)

print("✅ Pipeline executada com sucesso!")
print(df)

# Download do arquivo gerado
with open('data_sync.json', 'w') as f:
    f.write(result_json)
```
